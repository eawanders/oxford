```python
import pandas as pd
import random

# Load your dataset (replace 'your_file.xlsx' with your actual file)
file_path = "/Users/edwardanders/Desktop/Trinity Term 2025 - Exchanges Ballot.xlsx"
sheet_name = 'Sheet1'  # Replace with the correct sheet name if applicable

# Load the data into a DataFrame
df = pd.read_excel(file_path, sheet_name=sheet_name)

# Clean up column names by stripping any leading/trailing spaces
df.columns = df.columns.str.strip()

# Print out the column names to verify them
print("Columns in the DataFrame:", df.columns)

# Assuming column 'Name' contains names and ballot columns are specified by the event names
names_column = 'Name'
ballot_columns = ["St Hugh's", "Blackfriars", "Oriel", "Reuben", "Wolfson"]  # Event columns

# Number of spots per ballot
ballot_spots = {
    "St Hugh's": 14,
    "Blackfriars": 6,
    'Oriel': 14,
    "Reuben": 14,
    "Wolfson": 14
}

# Initialize an empty list to track successful candidates
successful_candidates = []

# Function to run a ballot using random selection and limit the number of spots
def run_ballot(df, column_name, spots_available, previous_winners):
    # Filter the DataFrame for candidates who answered 'Yes' and exclude previous winners
    eligible_candidates = df[df[column_name] == 'Yes']
    eligible_candidates = eligible_candidates[~eligible_candidates[names_column].isin(previous_winners)]

    # Sort alphabetically by names (to maintain consistent order)
    sorted_candidates = eligible_candidates.sort_values(by=[names_column])

    # Shuffle the sorted candidates
    shuffled_candidates = sorted_candidates.sample(frac=1).reset_index(drop=True)

    # Limit the selection to the number of available spots
    successful = shuffled_candidates[names_column].head(spots_available).tolist()

    return successful

# Loop over each ballot column and run the ballot
for ballot in ballot_columns:
    spots = ballot_spots[ballot]  # Get the number of spots for this event
    print(f"Running ballot for event: {ballot} with {spots} spots available.")

    # Run the ballot for the current column
    successful_this_round = run_ballot(df, ballot, spots, successful_candidates)

    # Print the results of the ballot
    print(f"Successful candidates for {ballot}: {successful_this_round}")

    # Update the list of successful candidates
    successful_candidates.extend(successful_this_round)

    # If no candidates are left, we can break early
    if not successful_this_round:
        print("No more eligible candidates.")
        break

print("All ballots complete!")
```
